```python
IMG_HEIGHT = 128
IMG_WIDTH = 256
MAX_SAMPLES = 1000

X = []
Y = []

for i, label in enumerate(labels[:MAX_SAMPLES]):
    image, mask = create_lane_mask(label, DATASET_PATH)

    image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT))
    mask = cv2.resize(mask, (IMG_WIDTH, IMG_HEIGHT))

    image = image / 255.0
    mask = mask / 255.0

    X.append(image)
    Y.append(mask)

    if i % 100 == 0:
        print(i, "adet veri hazırlandı")

X = np.array(X, dtype=np.float32)
Y = np.array(Y, dtype=np.float32)

Y = np.expand_dims(Y, axis=-1)

print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)

print("X_train:", X_train.shape)
print("Y_train:", Y_train.shape)
print("X_val:", X_val.shape)
print("Y_val:", Y_val.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model

def conv_block(inputs, filters):
    x = Conv2D(filters, 3, activation="relu", padding="same")(inputs)
    x = Conv2D(filters, 3, activation="relu", padding="same")(x)
    return x

def build_unet(input_shape=(128, 256, 3)):
    inputs = Input(input_shape)

    # Encoder
    c1 = conv_block(inputs, 16)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = conv_block(p1, 32)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = conv_block(p2, 64)
    p3 = MaxPooling2D((2, 2))(c3)

    # Bottleneck
    c4 = conv_block(p3, 128)

    # Decoder
    u5 = Conv2DTranspose(64, 2, strides=(2, 2), padding="same")(c4)
    u5 = concatenate([u5, c3])
    c5 = conv_block(u5, 64)

    u6 = Conv2DTranspose(32, 2, strides=(2, 2), padding="same")(c5)
    u6 = concatenate([u6, c2])
    c6 = conv_block(u6, 32)

    u7 = Conv2DTranspose(16, 2, strides=(2, 2), padding="same")(c6)
    u7 = concatenate([u7, c1])
    c7 = conv_block(u7, 16)

    outputs = Conv2D(1, 1, activation="sigmoid")(c7)

    model = Model(inputs, outputs)
    return model

model = build_unet()
model.summary()
```

```python
model = build_unet()

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=7,
    restore_best_weights=True
)

checkpoint = ModelCheckpoint(
    "/content/drive/MyDrive/lane_project/best_lane_unet_model.keras",
    monitor="val_loss",
    save_best_only=True
)

history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=100,
    batch_size=8,
    callbacks=[early_stop, checkpoint]
)
```

```python
def get_lane_status_from_mask(binary_mask, threshold=20):
    height, width = binary_mask.shape

    # Görüntünün alt yarısı daha önemli, çünkü araç konumu alt kısma göre yorumlanır
    lower_half = binary_mask[height // 2:, :]

    # Beyaz piksellerin koordinatlarını bul
    ys, xs = np.where(lower_half > 0)

    if len(xs) == 0:
        return "Lane Not Detected", None, width // 2, 0

    # Görüntü merkezini araç merkezi kabul ediyoruz
    car_center = width // 2

    # Sol ve sağ şerit piksellerini ayır
    left_xs = xs[xs < car_center]
    right_xs = xs[xs >= car_center]

    if len(left_xs) == 0 or len(right_xs) == 0:
        return "Lane Partially Detected", None, car_center, 0

    left_lane_x = int(np.mean(left_xs))
    right_lane_x = int(np.mean(right_xs))

    lane_center = (left_lane_x + right_lane_x) // 2

    deviation = car_center - lane_center

    if abs(deviation) <= threshold:
        status = "Center"
    elif deviation > threshold:
        status = "Right Departure"
    else:
        status = "Left Departure"

    return status, lane_center, car_center, deviation
```

```python
status, lane_center, car_center, deviation = get_lane_status_from_mask(binary_mask)

print("Lane Status:", status)
print("Lane Center:", lane_center)
print("Car Center:", car_center)
print("Deviation:", deviation)
```

```python
def draw_lane_status(image, binary_mask, status, lane_center, car_center, deviation):
    output = (image * 255).astype(np.uint8).copy()

    height, width = binary_mask.shape

    # Maske üstünü yeşilimsi bindirmek yerine beyaz pikselleri doğrudan gösterelim
    mask_rgb = np.zeros_like(output)
    mask_rgb[binary_mask > 0] = [255, 255, 255]

    overlay = cv2.addWeighted(output, 0.8, mask_rgb, 0.6, 0)

    # Araç merkezi çizgisi
    cv2.line(overlay, (car_center, 0), (car_center, height), (255, 0, 0), 2)

    # Şerit merkezi çizgisi
    if lane_center is not None:
        cv2.line(overlay, (lane_center, 0), (lane_center, height), (0, 255, 0), 2)

    cv2.putText(
        overlay,
        f"Lane Status: {status}",
        (10, 25),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.7,
        (255, 255, 255),
        2
    )

    cv2.putText(
        overlay,
        f"Deviation: {deviation}",
        (10, 55),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.7,
        (255, 255, 255),
        2
    )

    return overlay
```

```python
def predict_lane_status(image, model, threshold_value=0.5, departure_threshold=20):
    """
    image: 128x256x3 normalize edilmiş görüntü
    model: eğitilmiş U-Net modeli
    threshold_value: maskeyi binary hale getirme eşiği
    departure_threshold: merkez sapması için karar eşiği
    """

    # Model tahmini
    prediction = model.predict(np.expand_dims(image, axis=0), verbose=0)[0]
    pred_mask = prediction.squeeze()

    # Binary mask
    binary_mask = (pred_mask > threshold_value).astype(np.uint8) * 255

    # Şerit durumunu hesapla
    status, lane_center, car_center, deviation = get_lane_status_from_mask(
        binary_mask,
        threshold=departure_threshold
    )

    return status, binary_mask, lane_center, car_center, deviation
```
